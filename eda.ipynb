```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd=pd.read_csv("/train.csv")
pd.head(8)
pd.tail(8)


```

```python
pd.drop(["Cabin","SibSp"],axis=1)
```

```python
pd.rename(columns={"passenger":"psg"})
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(5,4))
sns.boxplot(data=pd[["Age","Fare"]],palette="muted")
plt.title("boxplot for outlier detetion")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Survied",data=pd,palette="pastel")
plt.title("Survival count",fontsize=14)
plt.show()
```

# New Section

```python
plt.figure(figsize=(6,5))
plt.pie(x="Fare",data=pd)
plt.title("piechart",fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x=,data=pd)
plt.title("scatterplot",fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(pd["Age"],bins=30,kde=True,color="blue")
plt.title("Age distribution",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Sex",y="Survived",data=pd,palette="coolwarm")
plt.title("survival rate by gender",fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.lineplot(x="Age",y="Sex",data=pd)
plt.title("linechart",fontsize=10)
plt.show()
```
